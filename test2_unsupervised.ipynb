Principle component analysis

```python
df_numeric_features = df_nooutlier.select_dtypes(include=[np.number])
```

```python
df_numeric_features.head()
```

```python
from sklearn.preprocessing import StandardScaler
df_num_std = StandardScaler().fit_transform(df_numeric_features)

print(df_num_std)
```

```python
cov_mat = np.cov(df_num_std.T)
cov_mat
```

```python
eig_val, eig_vec = np.linalg.eig(cov_mat)

print('Eigenvalues:','\n','\n', eig_val,"\n")

print('Eigenvectors:','\n','\n',eig_vec,'\n')
```

```python
eig_val = list(eig_val)
eig_val.sort(reverse = True) 
print(eig_val)
```

```python
plt.plot(eig_val)            

plt.plot(eig_val, "bp") 

plt.xlabel('Principal Components')    

plt.ylabel('Percentage of explained variance')      


plt.annotate('Elbow Point', xy=(4,1.5), xytext=(5, 2.5), arrowprops=dict(facecolor='black', arrowstyle = 'simple'))

# title of the plot
plt.title('Scree Plot')

plt.show()   
```

```python
eigenvector = eig_vec[:,0:4] 
eigenvector
```

```python
df_pca = pd.DataFrame(df_num_std.dot(eigenvector), columns= ['PC1','PC2','PC3','PC4']) 

df_pca.head()
```

```python
df_pca.shape
```

K Means clustering

```python
features_scaled = pd.DataFrame( df_num_std, columns= df_numeric_features.columns) #List of column names from the original dataframe data.columns[1:36]
features_scaled.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

n_clusters = [4, 5, 6, 7, 8] 
for K in n_clusters :
    cluster = KMeans (n_clusters= K, random_state= 10) 
    predict = cluster.fit_predict(features_scaled) 
    
    score = silhouette_score(features_scaled, predict, random_state= 10)
    print ("For n_clusters = {}, silhouette score is {})".format(K, score))
```

```python
# building  model for K = 5
model = KMeans(n_clusters= 5, random_state= 10)
model.fit(features_scaled)
```

```python
data_output = df_numeric_features.copy(deep = True)
data_output['Cluster'] = model.labels_ 
data_output.head()
```

```python
np.unique(model.labels_, return_counts=True)   
```

```python
sns.countplot(data= data_output, x = 'Cluster')


plt.title('Cluster Sizes', fontsize = 15)
plt.xlabel('Clusters', fontsize = 15)
plt.ylabel('No. of Customers', fontsize = 15)

plt.text(x = -0.18, y =2000, s = np.unique(model.labels_, return_counts=True)[1][0])
plt.text(x = 0.9, y =2000, s = np.unique(model.labels_, return_counts=True)[1][1])
plt.text(x = 1.85, y =2000, s = np.unique(model.labels_, return_counts=True)[1][2])
plt.text(x = 2.85, y =2000, s = np.unique(model.labels_, return_counts=True)[1][3])

plt.show()
```
